Importing the necessary libraries


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

The following concepts have been used:

    1.Reading the image
    2.Cropping the image to required area
    3.Converting to grayscale 
    4.Blurring the images
    5.Edge detection
    6.Dilation
    7.Making the images as per requiremnet

reading and cropping the images

```python
img1 =cv2.imread("1.png")
cropped_image1 = img1[200:600, 150:550]#cropping the required regions

img2 =cv2.imread("2.png")
cropped_image2 = img2[300:600, 0:500]

img3 =cv2.imread("3.png")
cropped_image3 = img3[250:550, 200:550]

img4 =cv2.imread("4.png")
cropped_image4 = img4[300:600, 0:500]
```

Using functions to perform rest of the concepts

```python
def image(img):
    
    imgGray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #CONVERT TO GRAY SCALE
    img_blur = cv2.GaussianBlur(imgGray, (3,3), 0) #BLUR USING GAUSSIAN BLUR
    #img_blur=cv2.medianBlur(imgGray,5) #BLUR USING MEDIAN BLUR
    edges = cv2.Canny(img_blur,25,25) #EDGE DETECTION
    kernel = np.ones((5, 5), np.uint8) #DEFINING THE KERNEL TO BE USED FOR DILATION
    img_dilation = cv2.dilate(edges, kernel, iterations=1) #DILATION
    imagem = cv2.bitwise_not(img_dilation) #TO MAKE IT AS ASKED FOR
    cv2.imwrite("final.png", imagem) 
    cv2.imshow("img",imagem)
    cv2.waitKey(0)
```

```python
image(cropped_image1)

```
